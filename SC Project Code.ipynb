```python
##Split the training and test sets
from sklearn.model_selection import train_test_split
x_train,x_validate,y_train,y_validate=train_test_split(X_train,Y_train,test_size=0.2,random_state = 12345)
print(x_train.shape)
print(y_train.shape)
print(x_validate.shape)
print(y_validate.shape)
```

```python
#reshape into 28*28*1
x_train=np.array(x_train).reshape(-1,28,28,1)
x_validate=np.array(x_validate).reshape(-1,28,28,1)
X_test = np.array(X_test).reshape(-1,28,28,1)
```

# Convolutional Neural Network

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
model=Sequential()

model.add(Conv2D(32, (3, 3), input_shape = (28,28,1), activation='relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), input_shape = (28,28,1), activation='relu')) #convolutional layer
model.add(MaxPooling2D(pool_size = (2, 2)))#maxpooling 
model.add(Dropout(0.25)) # to overcome overfitting

model.add(Conv2D(128, (3, 3), input_shape = (28,28,1), activation='relu')) 
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
#adding fully connected layer
model.add(Dense(units = 512, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(units = 25, activation = 'softmax'))
```

```python
model.summary()
```

```python
#compiling
model.compile(loss ='sparse_categorical_crossentropy', optimizer='adam' ,metrics =['accuracy'])
```

```python
#train the model
history = model.fit(x_train, y_train, batch_size = 512, epochs = 20, verbose = 1, validation_data = (x_validate, y_validate))
```

```python
#Visualizing the training performance
plt.figure(figsize=(12, 8))


plt.plot(history.history['loss'], label='Loss')
plt.plot(history.history['val_loss'], label='val_Loss')
plt.legend()
plt.grid()
plt.title('Loss evolution')
```

```python
plt.figure(figsize=(12, 8))
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.legend()
plt.grid()
plt.title('Accuracy evolution')
```

```python
#loss and accuracy for testing dataset
test_loss,test_acc=model.evaluate(X_test,Y_test)
print('Test Loss is: ',test_loss)
print('Test Accuracy is: ',test_acc)
```

As we can see in the above visualization, the CNN model has predicted the correct class labels for almost all the images. Now we will see the full classification report using a normalized and non-normalized confusion matrices. 

```python
import skfuzzy as fuzz
import skfuzzy.control as ctrl
import numpy as np

# CNN predictions for all test samples (probabilities)
cnn_probabilities = model.predict(X_test)

# Define fuzzy inputs for all 25 letters
letters = [chr(i) for i in range(65, 65 + 25)]  # Generate A-Y
fuzzy_inputs = {letter: ctrl.Antecedent(np.arange(0, 1.1, 0.1), f'prob_{letter}') for letter in letters}

# Define fuzzy outputs for final confidence percentages
fuzzy_outputs = {letter: ctrl.Consequent(np.arange(0, 101, 1), f'output_{letter}') for letter in letters}

# Membership functions for inputs (class probabilities)
for letter in letters:
    fuzzy_inputs[letter]['low'] = fuzz.trimf(fuzzy_inputs[letter].universe, [0, 0, 0.5])
    fuzzy_inputs[letter]['high'] = fuzz.trimf(fuzzy_inputs[letter].universe, [0.5, 1, 1])

# Membership functions for outputs (confidence percentages)
for letter in letters:
    fuzzy_outputs[letter]['low'] = fuzz.trimf(fuzzy_outputs[letter].universe, [0, 0, 50])
    fuzzy_outputs[letter]['high'] = fuzz.trimf(fuzzy_outputs[letter].universe, [50, 100, 100])

# Create fuzzy rules for decision-making
rules = []
for letter in letters:
    rules.append(ctrl.Rule(fuzzy_inputs[letter]['high'], fuzzy_outputs[letter]['high']))
    rules.append(ctrl.Rule(fuzzy_inputs[letter]['low'], fuzzy_outputs[letter]['low']))

# Create fuzzy control system and simulation
fuzzy_system = ctrl.ControlSystem(rules)
classifier = ctrl.ControlSystemSimulation(fuzzy_system)

# Process predictions for each test sample
for i, probs in enumerate(cnn_probabilities):
    # Assign CNN probabilities to fuzzy system inputs
    for j, letter in enumerate(letters):
        classifier.input[f'prob_{letter}'] = probs[j]

    # Compute fuzzy output
    classifier.compute()

    # Get fuzzy outputs for all letters
    fuzzy_result = {letter: classifier.output[f'output_{letter}'] for letter in letters}

    # Sort and display top 2 results
    sorted_result = sorted(fuzzy_result.items(), key=lambda x: x[1], reverse=True)
    top_predictions = sorted_result[:2]  # Top 2 predictions
    print(f"Sample {i + 1} Fuzzy Prediction: {top_predictions[0][1]:.2f}% {top_predictions[0][0]} and {top_predictions[1][1]:.2f}% {top_predictions[1][0]}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Initialize the true and predicted labels
true_labels = Y_test
fuzzy_predicted_classes = []

# Generate predictions from the fuzzy system
for probs in cnn_probabilities:
    # Assign CNN probabilities to fuzzy system inputs
    for j, letter in enumerate(letters):
        classifier.input[f'prob_{letter}'] = probs[j]

    # Compute fuzzy output
    classifier.compute()

    # Get fuzzy outputs for all letters
    fuzzy_result = {letter: classifier.output[f'output_{letter}'] for letter in letters}

    # Select the class with the highest fuzzy confidence
    predicted_class = max(fuzzy_result, key=fuzzy_result.get)
    fuzzy_predicted_classes.append(letters.index(predicted_class))

# Create the confusion matrix
cm = confusion_matrix(true_labels, fuzzy_predicted_classes)

# Normalize the confusion matrix
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

# Plot the confusion matrix
plt.figure(figsize=(20, 20))
sns.heatmap(cm_normalized, annot=True, cmap='Blues', xticklabels=letters, yticklabels=letters, fmt=".2f")
plt.xlabel('Predicted Label', fontsize=20)
plt.ylabel('True Label', fontsize=20)
plt.title('Normalized Confusion Matrix', fontsize=25)
plt.show()
```
